```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/cancer_train.csv')
df.head()
```

```python
x = df.drop('diagnosis', axis=1).values
t = df['diagnosis'].values
df.info()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, t_train, t_test = train_test_split(x,t, test_size=0.3, random_state=0)

from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
std_scaler.fit(x_train)
x_train_std = std_scaler.transform(x_train)
x_test_std = std_scaler.transform(x_test)
from sklearn.svm import SVC
svc = SVC()
svc.fit(x_train_std, t_train)
print('train score : ', svc.score(x_train_std, t_train))
print('test score : ', svc.score(x_test_std, t_test))
```

```python
test = pd.read_csv('/content/cancer_test.csv')
test.head()
```

```python
test_v = test.values
test_v
```

```python
test_std = std_scaler.transform(test_v)
y = svc.predict(test_std)
```

```python
test['pred'] = y
test.head()
```

```python
test['pred'] = test['pred'].map({'M':1, 'B':0})
test.head()
```

```python
y = test['pred'].values
y
```

```python
pd.Series(y, name='pred').to_csv('submission.csv', index = None)
```
