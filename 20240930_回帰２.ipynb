```python
for column in df.columns:
    if column != 'age':  # 'age' 列との相関は不要なので除外
        correlation = df['age'].corr(df[column])
        print(f"Column: {column}, Correlation with age: {correlation}")
```

```python
import pandas as pd

# 相関行列の計算
correlation_matrix = df.corr()

# 'age' 列との相関が高い列を抽出
age_corr = correlation_matrix["age"].sort_values(ascending=False)

# 結果を表示
print(age_corr)
```
