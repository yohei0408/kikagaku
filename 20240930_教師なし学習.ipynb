```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/pokemon_data.csv')
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = df.drop('Name',axis = 1)
x_scaled = scaler.fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca =PCA(n_components=2, random_state=0)
pca.fit(x_scaled)
x_transformed = pca.fit_transform(x_scaled)
pd.DataFrame(x_transformed, columns=['1', '2']).head(10)
```

```python
print('1:{}'.format(pca.explained_variance_ratio_[0]))
print('2:{}'.format(pca.explained_variance_ratio_[1]))
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(x)
```

```python
kmeans.cluster_centers_.shape
```

```python
cluster = kmeans.predict(x)
cluster
```

```python
df_cluster = df.copy()
df_cluster['cluster'] = cluster
df_cluster.head()
```

```python
df_results = pd.DataFrame()
df_results
```

```python
df_cluster[df_cluster['cluster'] == 0].max().tolist()
```

```python
# 各クラスターの平均値を追加
df_results['cluster 0'] = df_cluster[df_cluster['cluster'] == 0].max().tolist()
df_results['cluster 1'] = df_cluster[df_cluster['cluster'] == 1].max().tolist()
df_results['cluster 2'] = df_cluster[df_cluster['cluster'] == 2].max().tolist()
df_results
```

```python
df_results = df_results.set_index(df_cluster.columns)
df_results = df_results.drop(['Name','cluster']).T
df_results
```

```python
# 'cluster' が 1 の行をフィルタリングし、'HP' の最大値を取得
max_hp_cluster_1 = df_cluster[df_cluster['cluster'] == 1]['HP'].max()

# 結果を表示
print(f"cluster が 1 の 'HP' の最大値: {max_hp_cluster_1}")
```
